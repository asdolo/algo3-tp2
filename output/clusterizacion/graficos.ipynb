```python
#TEST 1 Complejidades Clusterizacion Version 1
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)

primTest1['FB/N^3'] = primTest1['tiempo'] / primTest1['n']**3;
ax1 = sns.lineplot(x="n", y="FB/N^3", data=primTest1.reset_index(), label="Prim", ax=ax);
kruskalTest1['FB/N^3'] = kruskalTest1['tiempo'] / kruskalTest1['n']**3;
ax1 = sns.lineplot(x="n", y="FB/N^3", data=kruskalTest1.reset_index(), label="Kruskal", ax=ax);
kruskalPathCompressionTest1['FB/N^3'] = kruskalPathCompressionTest1['tiempo'] / kruskalPathCompressionTest1['n']**3;
ax1 = sns.lineplot(x="n", y="FB/N^3", data=kruskalPathCompressionTest1.reset_index(), label="Kruskal-Path-Compression", ax=ax);
#Reverse legends
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(reversed(handles), reversed(labels), title='LÃ­neas', loc='upper right')

plt.ylim(-0.005,0.005);

plt.ylabel(r"F(N,V) = $\frac{f(N, V)}{N^3}$", fontsize=16);
plt.xlabel("n", fontsize=16);
plt.tick_params(labelsize=14)
plt.setp(ax.lines,linewidth=3)
```
